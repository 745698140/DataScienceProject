```python
import matplotlib.pyplot as plt

import numpy as np

x = [1, 50, 500, 5000, 25000]
y = [1.667, 0.568, 0.056, 0.006, 0.007]

plt.plot(x, y)
plt.xlabel("The num of work")
plt.ylabel("Response time of each work(s)")
#在ipython的交互环境中需要这句话才能显示出来
plt.show()
```
