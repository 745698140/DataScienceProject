```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
names=['CRIM','ZN','INDUS','CHAS','NOX','RM','GE','DIS','RAD','TAX','PRTATIO','B','LSTAT','PRICE']

boston=pd.read_csv("housing.csv")
boston.head(10)
```

```python
corrboston = boston.corr()
corrboston

plt.figure(figsize=(10,10))    
sns.heatmap(corrboston,annot=True,cmap='RdGy')
plt.show()
```

```python
fig,ax = plt.subplots(1,2,figsize=(10,5))

boston['CHAS'].value_counts().plot.pie(ax=ax[0],shadow=False,autopct='%1.2f%%')
ax[0].set_ylabel('')   
ax[0].set_xlabel('CHAS')    
sns.countplot('CHAS',data=boston,ax=ax[1])
ax[1].set_ylabel('')
ax[1].set_xlabel('CHAS')
plt.show()
```

```python

bostonCHAS = boston[['CHAS','MEDV']]    

bostonCHAS1=bostonCHAS.pivot_table(values='MEDV',    
                               index='CHAS',       
                               aggfunc='mean')           


bostonCHAS1 = bostonCHAS1.sort_values(by='MEDV',     
                        ascending=False                 
                       )

bostonCHAS1
```

```python
x_data = boston[['CRIM','ZN','INDUS','CHAS','NOX','RM','AGE','DIS','RAD','TAX','PTRATIO','B','LSTAT']]
y_data = boston[['MEDV']] 

plt.figure(figsize=(18,10))

for i in range(13):
    plt.subplot(4,4,i+1)
    plt.scatter(x_data.values[:,i],y_data,s = 5)    #.values将DataFrame对象X_df转成ndarray数组
    plt.xlabel(names[i])
    plt.ylabel('MEDV')
    plt.title(str(i+1)+'. '+names[i]+' - MEDV')  
    
plt.tight_layout()
plt.show()
```
